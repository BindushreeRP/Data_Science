```python
#Text adventure game by Bindu
#Source of this game is from the movie Arya 2: https://en.wikipedia.org/wiki/Arya_2

input(prompt="""Hello! Welcome to the text adventure game..
You will go through a series of questions/levels to reach the final stage
based on the Indian movie Arya 2!
Press the 'enter' key to continue.. \n""")

#Printing the introduction before the game, inspired by the movie Arya2 in Telugu

print(f"""In this game you will experience the life of Arya, he's rebellious,
charming adult, that falls in love with the same girl as his best friend...
He and his friend, Ajay met as kids in an orphanage, where Arya being a rebel
forced his friendship upon Ajay. Ajay was adopted into a wealthy family
while Arya grew up at the orphanage. Ajay is now a multimillionaire 
businessman. Upon Arya's pleading, he hires Arya to be in one of his teams.
They both see Geetha a new hire at Ajay's firm and are instantly in love. 
It is now what enfolds between these three adults that you will see as part 
of this game... Hope you enjoy it!""")

#Importing sys and random
import sys
import random

#Defining fail in order to use them in the stages further.

def fail():
    """This function is used when the user runs out of chances or inputs the wrong answers"""
    print(f"""You have failed!! Bad luck.. :(""")
    sys.exit()

#Defining scene_1 and its purpose in the docstring
def scene_1():

#doc string for scene_1

    """This is the first scene to put forth the characteristics of each character in this movie. 
       This scene consists of various functions such as random choice, lists being defined, for and
       while loops with conditional and nested conditional loops."""

# setting up list_1 with the answers to questions
    list_1 = [
'1 He made Ajay eat meat', 
'2 He forced Ajay to get his name, Arya tattooed',
'3 He pushed Ajay into a water pond to test their friendship']

#Created list with the options to use in the random choice 
    list_2 = ['1','2','3']
    list_answer = random.choice(list_2)

#question for scene 1    
    print("""
Here is a question to see if you qualify for the next round!
Which one of the below options is true about Arya?""")

#creating for loop to print list options.
    for options in list_1:
        print(options)

#defining the number of chances for this question
    chances = 2
    
#using while loop to incorporate the chances 
    while chances > 0:
        choice = input(prompt="Insert the number to right option from above: \n""").lower()

#start of a conditional statement within the while loop

        if choice == list_answer:
            print(f"""That's correct. You are off to the next stage. \n""")
            break
    
        elif choice != list_answer:
            chances -= 1
        
#start of nested conditional if chances were available        
            if chances > 0:
                print(f"""That's incorrect.. You are left with {chances} chance. \n""")
        
            else:
                print(f"""You are out of chances! Right answer was option {list_answer}""")
                fail()
            
        elif choice != list_answer:
            print(f"""Answer is {list_answer}""")
            chances -= 1
        
#2nd nested conditional loop if any chances were available
            if chances > 0:
                print(f"""That's incorrect.. You are left with {chances} chance. \n""")
        
            else:
                print(f"""You are out of chances. Right answer was option {list_answer}\n""")
                fail()
        
        else:
            print("""Something went wrong, there was a bug in your input. \n""")
            fail()
scene_1()

#defining scene 2 and building the game

def scene_2():

#setting up docstring for scene_2
    """This scene gives a little insight into the movie and its characters"""
    """Scene 2 dives into more specifics of what happens between the characters"""

#setting a variable like a madlib, actual world would be 'Kisses/kissed'!
    action = input(prompt="""
Enter a romantic action verb here..>\n""").capitalize()

#Defining options to questions in this scene as a list 2 and 3

    list_2 = ['1 Breaks her phone',
              '2 Screams loudly',
              '3 Bangs the door']
    
    list_3 = ['True', 'False']

#printing the resulting madlib statement

    print(
f"""
Arya is in a lift, Geetha enters the same lift and he {action} her!!
""")

#printing the question for scene 2..

    print(f"""
Geetha was stunned by Arya's action and she has a habit of doing something when
she's very angry.. Guess the habit from the below options: """)
    
#setting up for loop to print the options
    for habit in list_2:
        print(habit)
        continue
        
#defining the input variable to receive the answer   
    answer = input(prompt="""Please enter the right answer: \n""").lower()

#start of the conditional statement

    if answer == '1' or 'one' in answer or answer == 'One' or 'ONE' in answer or answer == '1 Breaks her phone' or 'break' in answer:
        
        print(f"""
That's correct!! You may now answer the next question: Do you think(
Ajay will make use of Geetha's anger towards Arya to his advantage 
and get close to Geetha?""")
        
#for loop to print the options
        for value in list_3:
            print(value)
            continue

#defining the answer variable to the nested question    
        answer_2 = input(prompt="""Enter true or false: \n""").lower()
        
#starting a nested conditional loop
        if answer_2 == 'true' or 'True' in answer_2 or answer_2 == 'TRUE' or 't' in answer_2 or 'yes' in answer_2:
            print(f"""That's the right answer""")
        
        elif answer_2 == 'false' or '' in answer_2 or answer_2 == 'f' or 'fal' in answer_2 or 'no' in answer_2:
            print(f"""That's incorrect! :( """)
            fail()
               
        else:
            print(f"""Sorry that's incorrect! :(""")
            
            fail()
    
    elif answer == '2' or 'two' in answer or answer == 'Two' or '' in answer or answer == '2 Screams loudly' or 'scream' in answer:
        print(f"""Sorry! That's incorrect..""")
        
        fail()
    
    elif answer == '3' or 'three' in answer or answer == 'Three' or '' in answer or answer == '3 Bangs the door' or 'bang' in answer:
        print(f"""Sorry! That's incorrect..""")
    
        fail()
    
    else:
        print("""Something went wrong. There was a bug in your input.""")

scene_2()
    
#Defining and setting up scene 3..

def scene_3():

#setting up the docstring for scene_3

    """Getting into the characters and this scene just uses a for loop and a nested conditional statement."""
    
    print(f"""
Arya has this habit of behaving badly or possessively with any guy that 
tries to get close to Geetha.. There comes a day when Geetha had to vacate
her apartment because Arya thought of Geetha' landlord as yet another guy 
trying to come close to Geetha!! She becomes homeless and blames Arya.. 
Arya arranges an apartment for Geetha.. \n""")

# defining the list with options to question in scene 3
    list_4 = ['No','Yes']

    print(f"""
Do you think Arya made Geetha stay at his apartment while 
he was secretly living there?""")

#instantiating for loop to print the options to above questions

    for yn in list_4:
        print(yn)
        continue
        
#defining the input variable
    
    answer_3 = input(prompt="""Enter your answer here: \n""").lower()
    
#starting conditional loop
    
    if answer_3 == 'yes' or 'Yes' in answer_3 or answer_3 == 'YES' or 'yeah' in answer_3 or 'y' in answer_3:
        print(f"""That is correct.\n""")
        
    elif answer_3 == 'no' or 'No' in answer_3 or answer_3 == 'NO' or answer_3 == '' or 'n' in answer_3:
        print(f"""That is incorrect.\n""")
        fail()
    
    else:
        print("""Something went wrong, try again!""")
        fail()

scene_3()

#defining and setting up scene 4

def scene_4():

#defining the doc string for scene_4
    """This scene consists of nested conditional statement and user input 
       manipulation is handled by the .lower function and having incorporated
       multiple possible inputs by users for both right and wrong answers."""
#printing the question and input prompt 

    print(f"""
Meanwhile, Geetha's father would have asked her to get to their home
at a small village in South India.. She calls Ajay and tells him that
her father has arranged her marriage with his friend's son called Subbi! \n""")
    
    print(input(prompt="""Press the 'enter' key to continue. \n"""))
    
    print(f"""
What do you think Ajay will do?
1 Go talk to Geetha's father
2 Reach Arya for help
3 Nothing
\n""")

#setting up input variable for the answer

    answer_4 = input(prompt="""Please enter the right answer from above: \n""").lower()

#starting conditional loop

    if answer_4 == '1' or 'one' in answer_4 or answer_4 == 'One' or 'ONE' in answer_4:
        print(f"""That is incorrect..""")
        fail()
    
    elif answer_4 == '2' or 'two' in answer_4 or answer_4 == 'Two' or 'TWO' in answer_4 or answer_4 == 'arya':
        print(f"""That is correct, you are on to the final stage""")
                
    elif answer_4 == '3' or 'three' in answer_4 or answer_4 == 'Three' or 'THREE' in answer_4 or answer_4 == 'nothing' or '' in answer_4:
        print(f"""That is incorrect..""")
        fail()
    
    else:
        print(f"""Something went wrong, try again""")
        fail()

scene_4()

#defining scene 5
def scene_5():

#defining the doc string for scene 5

    """This scene is the final scene, using random.choice to randomly pick a final character name.
       This scene has some nested conditional loop, random choice and user input variables that are defined."""
    
#printing the scene 
    print(f"""
Some events unfold, Geetha starts to question her choice about Ajay as
he always wanted Arya to solve his problems.. And then, there was Arya,
a rebel and Subbi who truly loves her.. \n""")
    
    print(f"""Among this chaos who do you think Geetha must choose as her partner?""")
    
#creating a list
    name = ['Arya','Ajay','Subbi']
    character = random.choice(name).lower()

#defining the user input variable
    user_input = input(prompt = """
Arya
Ajay
Subbi
\n""").lower()

#starting the conditional statement 
    if character == user_input:
        print(f"""That is correct, Geetha lives happily ever after with {character}!""")
    
    elif character != user_input or '' in user_input:
        print(f"""That is incorrect! :( Right answer was {character}""")
    
    else:
        print(f"""Something went wrong, try again later!""")

scene_5()   
    
#Ending the game with a print statement    
print(f"""
------------------------------------------------------------------------------
The End!
If curious to know more, please watch this movie available on YouTube.
------------------------------------------------------------------------------
""")
    
```
